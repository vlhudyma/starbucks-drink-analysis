```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv('starbucks_drinkMenu_expanded.csv')
```

```python
file.head()
```

```python
file.info()
```

```python
for column in file:
    count = file[column].apply(lambda x: isinstance(x, str) and any(c.isalpha() for c in x)).sum()
    percent = (count / len(file)) * 100
    print(f"{column}: {percent:.2f}%")
```

```python
change_t = file.columns[3:]

for column in change_t:
    file[column] = pd.to_numeric(file[column], errors='coerce')
    
```

```python
file.info()
```

```python
file.isnull().sum()
```

```python
file.drop(file.columns[13:17], axis = 1, inplace = True)
```

```python
calories_array = file['Calories'].to_numpy()
calories_array
```

```python
print(round(np.min(calories_array), 1), 
      round(np.max(calories_array), 1), 
      round(np.mean(calories_array), 1))
```

```python
calories_kj = np.round(calories_array * 4.184, 1)
calories_kj
```

```python
file['Calories kj'] = calories_kj
file
```

```python
file.to_csv("Starbucks_project.csv", index = True, encoding ='utf-8')
```

```python
# Analytical Insights

# 1. Find the drink with the highest number of calories
the_most_calories = file['Calories'].idxmax()
```

```python
file.loc[[the_most_calories]]
```

```python
# 2. Find the drink with the highest amount of caffeine
the_most_caffeine = file['Caffeine (mg)'].idxmax()
file.loc[[the_most_caffeine]]
```

```python
# 3. Compare if it's the same drink
if the_most_calories == the_most_caffeine:
    print('The drink with the most calories also contains the most caffeine.')
else:
    print('The drink with the most calories does not contain the most caffeine.')
```

```python
# Calculate the average sugar content per beverage category
group = file.groupby('Beverage_category')['Sugars (g)'].mean()
category_index = group.idxmax() 
print(f"The sweetest category is '{category_index}' with an average of {group[category_index]:.2f} grams of sugar.")
```

```python
sorted_file = file.sort_values(by="Sugars (g)", ascending=False)
top5_sugary = sorted_file[['Beverage', 'Beverage_category', 'Sugars (g)']].head(5)
top5_sugary
```
